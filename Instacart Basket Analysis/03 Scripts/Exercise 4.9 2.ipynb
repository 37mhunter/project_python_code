```python
#import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#create path
path=r'C:\Users\37mhu\Documents\Instacart Basket Analysis'
```

```python
#import custs_ords_prods data
custs_ords_prods=pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','custs_ords_prods.pkl'))
```

```python
#2 create first set of visualizations
#create bar chart 
bar_orders_dow=custs_ords_prods['orders_day_of_week'].value_counts().sort_index().plot.bar(title='Orders by Day of Week' )
```

```python
#remove price outliers
custs_ords_prods.loc[custs_ords_prods['prices'] >100, 'prices'] = np.nan
```

```python
#create scatterplot
scatterplot_prices=sns.scatterplot(x = 'prices', y = 'prices',data = custs_ords_prods)
```

```python
#create histogram
hist_prices=custs_ords_prods['prices'].plot.hist(title='Order Price Frequency',xlabel='price', ylabel='number of orders (in 10s of millions)', bins=25)
```

```python
#create line graph
np.random.seed(4)
dev = np.random.rand(len(custs_ords_prods)) <= 0.7
```

```python
big = custs_ords_prods[dev]
small = custs_ords_prods[~dev]
```

```python
sample=small[['orders_day_of_week','prices']]
```

```python
line_order_price = sns.lineplot(data = sample, x = 'orders_day_of_week',y = 'prices')
```

```python
#3 create histogram for hour of day column
hist_orders=custs_ords_prods['hour_order_placed'].plot.hist(title='Amount of Orders Placed by Hour', ylabel='number of orders (in 10s of millions)', xlabel='hour',bins=24)
```

Orders peak around the 10th hour, start to decline after the 15th hour, and reach their lowest around the 3rd hour

```python
#4 create bar chart for loyalty flag column
bar_loyalty=custs_ords_prods['loyalty_flag'].value_counts().plot.bar(title='Customer Loyalty', ylabel='user count (in 10s of millions)')
```

```python
#5 create line chart comparing prices to what hour an order was placed
sample2=small[['hour_order_placed','prices']]
```

```python
line_hour_order = sns.lineplot(data = sample2, x = 'hour_order_placed',y = 'prices')
```

Order prices are highest around the 3rd hour and lowest around the 10th hour.

```python
#6 create line chart comparing age to number of dependants
sample3=small[['Age','dependants']]
```

```python
line_age_dependants = sns.lineplot(data = sample3, x = 'Age',y = 'dependants')
```

There doesn't appear to be any relation between age and number of dependants.

```python
#7 create scatterplot comparing age to income
scatter_age_income=sns.scatterplot(x = 'Age', y = 'income',data = custs_ords_prods)
```

There doesn't appear to be a relationship between age and income, except that incomes over $400,000 appear to be more common after the age of 40.
